```python
x=np.arange(0,3.14*2,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x=np.arange(0,3.14*2,0.1)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1, label='sine')
plt.plot(x,y2,label='cosine')
plt.title('sine & cosine graph')
plt.grid(True)
plt.legend()
plt.show()
```

```python
x=np.linspace(-np.pi,np.pi,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1, label='sine')
plt.plot(x,y2,label='cosine')
plt.title('sine & cosine graph')
plt.grid(True)
plt.legend()
plt.show()
```
