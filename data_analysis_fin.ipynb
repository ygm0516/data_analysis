```python
#데이터 살펴보기

#데이터의 행x열 개수,컬럼,인덱스 정보 보기
CCTV_Seoul.shape

#데이터의 위에서 5행 보기
CCTV_Seoul.head(5)

#데이터의 아래에서 5행 보기
CCTV_Seoul.tail(5)

#데이터의 통계정보 보기
CCTV_Seoul.describe()

#데이터의 타입, 아이템 수, 누락데이터 등 보기
CCTV_Seoul.info()
```

```python
#데이터 정리하기

CCTV_Seoul.rename(columns={'구분':'구별'},inplace=True) #칼럼 이름 변경

CCTV_Seoul.drop(columns='2020년',axis=1,inplace=True) #불필요한 칼럼 삭제

#칼럼 추가
CCTV_Seoul['2017년 이전']=CCTV_Seoul['2011년 이전']+CCTV_Seoul['2011년']+CCTV_Seoul['2012년']+CCTV_Seoul['2013년']+CCTV_Seoul['2014년']+CCTV_Seoul['2015년']+CCTV_Seoul['2016년']

#불필요한 열들 삭제 (복수)
CCTV_Seoul.drop(columns=['2011년 이전', '2011년', '2012년', '2013년', '2014년','2015년','2016년'], axis=1,inplace=True)
CCTV_Seoul.head()
```

```python
#불필요한 행 삭제

CCTV_Seoul.drop(index=0, axis=0,inplace=True)
CCTV_Seoul.head()
```

```python
#drop = True 기존의 인덱스는 버리고 새로 설정
CCTV_Seoul.reset_index(drop=True, inplace=True)
CCTV_Seoul.head()
```

```python
#데이터 자세히 보기

#데이터 '총계'열을 기준으로 오름차순 정렬하기 (by='총계', ascending = True)
CCTV_Seoul.sort_values(by='총계', ascending = True)

#데이터 '총계'열을 기준으로 내림차순 정렬하기 (by='총계', ascending = False)
CCTV_Seoul.sort_values(by='총계', ascending = False)

#최근 3년간 CCTV 수를 더하고 2017년 이전 CCTV수로 나눠서 최근 3년간 CCTV 증가율 계산
CCTV_Seoul['최근증가율']=(CCTV_Seoul['2019년'] + CCTV_Seoul['2018년'] + CCTV_Seoul['2017년']) / CCTV_Seoul['2017년 이전']  * 100
CCTV_Seoul.rename(columns={'총계':'CCTV수'},inplace=True)

#'최근증가율'로 내림차순 정렬
CCTV_Seoul.sort_values(by='최근증가율', ascending = False)
```

```python
#서울시 데이터 읽어오기
#header= 1 옵션은 액셀에서 두번째 줄부터 읽어옴 (숫자는 0부터 시작)
pop_Seoul = pd.read_excel('/content/drive/MyDrive/대학교파일/4학년1학기/데이터분석/11주차 데이터 분석/행정구역_시군구_별__성별_인구수_서울.xlsx')
pop_Seoul.head()
```

```python
#데이터의 행,열 개수,컬럼,인덱스 정보 보기
print(pop_Seoul.shape)
print(pop_Seoul.index)
```

```python
#컬럼 이름 변경
pop_Seoul.rename(columns={'행정구역(시군구)별':'구별'}, inplace=True)
pop_Seoul.rename(columns={'2019. 12':'총인구수'}, inplace=True)
pop_Seoul.rename(columns={'2019. 12.1':'남자인구수'}, inplace=True)
pop_Seoul.rename(columns={'2019. 12.2':'여자인구수'}, inplace=True)
pop_Seoul.head()
```

```python
#불필요한 행 삭제, 0번 인덱스 줄 삭제
pop_Seoul.drop (index=0, axis=0, inplace=True)
pop_Seoul.head()
```

```python
#인덱스 새로 설정
pop_Seoul.reset_index(drop=True, inplace=True)
pop_Seoul.head()
```

```python
#'인구수'로 내림차순 정렬
pop_Seoul.sort_values(by='총인구수', ascending=False)
```

```python
#여성인구와 남성인구의 비율 알아보기
pop_Seoul['여성비율']=(pop_Seoul['여자인구수']/pop_Seoul['총인구수']*100).astype(int)
pop_Seoul['남성비율']=(pop_Seoul['남자인구수']/pop_Seoul['총인구수']*100).astype(int)

pop_Seoul.head()
```

```python
#'여성비율'로 내림차순 정렬
pop_Seoul.sort_values(by='여성비율', ascending=False).head(5)
```

```python
#'남성비율'로 내림차순 정렬
pop_Seoul.sort_values(by='남성비율', ascending=False).head(5)
```

```python
# 두 데이터 합치기 (병합기준 열을 '구별'로 병합, on='구별')
data_result=pd.merge(pop_Seoul,CCTV_Seoul, on='구별')

data_result.head()
```

```python
data_result.set_index('구별', inplace=True)
data_result.head()
```

```python
#합친 데이터 엑셀파일로 저장하기
# Colab에서는 파일의 경로명을 해당 파일 옆에 점 세개 버튼의 경로 복사로 복사해 사용
# 예) excel_writer = pd.ExcelWriter('/content/drive/MyDrive/Colab Notebooks/AI_2주차_실습파일_오프라인/CCTV_pop.xlsx')

excel_writer = pd.ExcelWriter('/content/drive/MyDrive/대학교파일/4학년1학기/데이터분석/11주차 데이터 분석/CCTV_pop.xlsx')
data_result.to_excel(excel_writer, index=True)
excel_writer.close()
```

```python
# 'CCTV비율' 열 추가
data_result['CCTV비율'] = data_result['CCTV수'] / data_result['총인구수'] * 100
data_result.head()
```

```python
# 소팅한 CCTV비율 막대 그래프 (kind='barh', grid=True, figsize=(10,10))
data_result['CCTV비율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
# 데이터에 대한 대표선을 나타내는, 기울기, 절편을 fp1에 저장
data_result['총인구수'] = data_result['총인구수'].astype(float)
data_result['CCTV수'] = data_result['CCTV수'].astype(float)
fp1 = np.polyfit(data_result['총인구수'], data_result['CCTV수'], 1)
fp1
```

```python
# 직선의 y값 생성시켜주는 함수
f1 = np.poly1d(fp1)
# 지정한 범위 내에서 데이터의 개수만큼 같은 간격의 숫자 생성, x축 숫자 생성
fx = np.linspace(100000, 700000, 100)
```

```python
# 산점도 그래프 그리기
plt.figure(figsize=(9,9))
plt.scatter(data_result['총인구수'],data_result['CCTV수'], s=50)

# 선그래프 그리기, 순서대로, x축 데이터, y축 데이터, lw는 선 굵기
plt.plot(fx, f1(fx), '--', lw=2, color='g')

#점 옆에 구이름 텍스트 넣기
for n in range(25):
    plt.text(data_result['총인구수'][n]*1.02, data_result['CCTV수'][n]*0.98, data_result.index[n], fontsize=10)

plt.xlabel('총인구수')
plt.ylabel('CCTV수')
plt.grid()
plt.show()
```
